```python
from numpy import*
from cmath import exp,pi
from pylab import*
data = loadtxt("sunspots.txt", float)
y=data[:,1]
g=data[:,1]
def dft(y):
    N = len(y)
    c = zeros(N//2+1,complex)
    for k in range(N//2+1):
        for n in range(N):
            c[k] += y[n]*exp(-2j*pi*k*n/N)
    return c
c=dft(y)
```

```python
from numpy import*
from cmath import exp,pi
from pylab import*

plot(abs(c))
xlim(-50,1600) 
xlabel('frecuencia')
ylabel('Ck')
show()
```

Escriba un programa para hacer lo siguiente;

a) Lea los datos de dovw. txt y trazarlos en un gráfico.

c) Ahora establezca todos menos el primer 10% de los elementos de esta matriz en cero (es decir, establezca el último 90% en cero pero mantenga los valores del primer 10%).

d) Calcule la transformada inversa de Fourier de la matriz resultante, ceros y todo, usando la función irfft, y grafíquela en la misma gráfica que los datos originales. Es posible que deba variar los colores de las dos curvas para asegurarse de que ambas aparezcan en el gráfico. Comenta lo que ves. ¿Qué sucede cuando configura los coeficientes de Fourier en cero?

# A

```python
y = loadtxt("dow.txt", float)
plot(y)
```

# B

```python
from numpy.fft import rfft
c = rfft(y)
```

# C

```python
a=[]
for n in range(len(c)):
    if n<len(c)//10:
        f=c[n]
        a.append(f)
    if n>=len(c)//10:
        f=c[n]*0
        a.append(f)
```

# D


```python
d= irfft(a)
plot(y)
plot(d)
```
